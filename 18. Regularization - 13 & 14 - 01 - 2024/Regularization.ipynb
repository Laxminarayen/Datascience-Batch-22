```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso,Ridge, ElasticNet
```

```python
df = pd.read_csv("test.csv",names = ['X','Y'])
```

```python
df
```

```python
model = LinearRegression()
model.fit(df[['X']],df[['Y']])
```

```python
pred = model.predict(df[['X']])
plt.scatter(df.X,df.Y,color='red')
plt.plot(df.X,pred)
```

```python
#lambda - alpha
model_ = Ridge(alpha = 0.005)
model_.fit(df[['X']],df[['Y']])

pred_ridge = model_.predict(df[['X']])

plt.scatter(df.X,df.Y,color='red')
plt.plot(df.X,pred)
plt.plot(df.X,pred_ridge,color = 'green')
```

```python
for i in np.arange(0,1,0.001):
  #lambda - alpha
  model_ = Ridge(alpha = i)
  model_.fit(df[['X']],df[['Y']])
  print("Slope: ",model_.coef_)
  print("intercept: ",model_.intercept_)
  pred_ridge = model_.predict(df[['X']])

  plt.scatter(df.X,df.Y,color='red')
  plt.plot(df.X,pred)
  plt.plot(df.X,pred_ridge,color = 'green')
```

```python
for i in np.arange(0,1,0.001):
  #lambda - alpha
  model_ = Lasso(alpha = i)
  model_.fit(df[['X']],df[['Y']])
  print("Slope: ",model_.coef_)
  print("intercept: ",model_.intercept_)
  pred_ridge = model_.predict(df[['X']])

  plt.scatter(df.X,df.Y,color='red')
  plt.plot(df.X,pred)
  plt.plot(df.X,pred_ridge,color = 'green')
```

```python
#lambda - alpha
model_ = ElasticNet(l1_ratio =0.2,alpha = 0.003)
model_.fit(df[['X']],df[['Y']])

pred_ridge = model_.predict(df[['X']])

plt.scatter(df.X,df.Y,color='red')
plt.plot(df.X,pred)
plt.plot(df.X,pred_ridge,color = 'green')
```
