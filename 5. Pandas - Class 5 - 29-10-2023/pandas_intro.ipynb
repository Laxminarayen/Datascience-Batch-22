## LOC And ILOC

```python
movies_df = pd.read_csv("IMDB-Movie-Data.csv")
```

```python
movies_df.head(6)
```

```python
movies_df.tail()
```

```python
movies_df.loc[3:7,['Title','Genre','Director']]
```

```python
#Everthing from my dataframe but using the concepts of list slicing
movies_df.loc[:,['Title','Genre']][3:7]
```

```python
movies_df.loc[3:7,3:6]
```

```python
movies_df.iloc[7,['Title','Genre']]
```

```python
movies_df.iloc[0:7,[1,2,3]]
```

```python
movies_df.loc[:,'Actors']
```

```python
movies_df.iloc[:,-2]
```

```python
list(movies_df.columns)
```
