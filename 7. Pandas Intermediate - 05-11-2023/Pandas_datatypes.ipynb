## Apply
Signature: df.apply(func, axis=0, broadcast=None, raw=False, reduce=None, result_type=None, args=(), **kwds)
Docstring:
Apply a function along an axis of the DataFrame.

Objects passed to the function are Series objects whose index is
either the DataFrame's index (``axis=0``) or the DataFrame's columns
(``axis=1``). By default (``result_type=None``), the final return type
is inferred from the return type of the applied function. Otherwise,
it depends on the `result_type` argument.

```python
lst= []
for i in df['Meal Price']:
    lst.append(i*i)
```

```python
lst
```

```python
df['Meal Squared'] = lst
```

```python
def square(x):
    return x*x
df['Meal Squared func'] = df['Meal Price'].apply(square)
```

```python
df['Meal Squared apply'] = df['Meal Price'].apply(lambda x : x*x)
```

```python
df.tail()
```

```python
df.loc[60000,'Meal Squared'] = "A String value"
```

```python
df['Meal Squared'] = pd.to_numeric(df['Meal Squared'], errors = 'coerce')
```

```python
df.tail()
```

```python
df['Meal Squared'] = df['Meal Squared'].fillna(df['Meal Squared'].mean())
```

```python
df['Meal Squared apply'] = df['Meal Squared apply'].fillna(df['Meal Squared apply'].median())
```

```python
df['Meal Squared'].astype(int)
```

```python
df['Meal Squared func'] = df['Meal Squared func'].fillna(0)
df['Meal Squared func'] = df['Meal Squared func'].replace(0,df['Meal Squared apply'].median())
```

## Adding rows or Combining dataframes

```python
new_row = {'Order Id':'3224frefdsfd3r24',
           'Date':'2016-04-22',
           'Meal Id':'In423fdsfsf'}
```

```python
df_test = df.copy()
```

```python
df_test = df_test.append(new_row,ignore_index=True)
```

```python
df_test.tail()
```

```python
new_row = {'Order Id':['3224frefdsfd3r24','frdsfdsf34434234'],
           'Date':['2016-04-22','2017-06-24'],
           'Meal Id':['In423fdsfsf','sdgsdgdsg']}
```

```python
add = pd.DataFrame(new_row)
```

```python
add
```

```python
df_new = pd.concat([df_test,add],ignore_index =True)
```

```python
df_new.tail()
```

```python
df_new_withoutignore = pd.concat([df_test,add])
```
