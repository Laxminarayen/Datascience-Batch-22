## Bar Plot

```python
flight_delay = pd.read_csv("data-for-datavis/flight_delays.csv",index_col="Month")
```

```python
flight_delay
```

```python
#Figure size setting
plt.figure(figsize = (20,10))

plt.title("Average flight delya fro UA flight, by month")
sns.barplot(x = flight_delay.index, y = flight_delay['UA'])

plt.ylabel("Arrival delay in (mins)")

plt.savefig("Bar.png",dpi=300)
```

## Heatmap

```python
#Figure size setting
plt.figure(figsize = (20,10))

plt.title("Heatmap of all the columns for flight delays")

sns.heatmap(data=flight_delay)

plt.xlabel("Airline Names")
```

```python
#Figure size setting
plt.figure(figsize = (20,10))

plt.title("Heatmap of all the columns for flight delays")


fligh_corr = flight_delay.corr()
sns.heatmap(data=fligh_corr,annot = True,vmin = -1,vmax = 1,center=0)

plt.xlabel("Airline Names")
```

## Scatter - Regplot

```python
insurance_data = pd.read_csv("data-for-datavis/insurance.csv") 
```

```python
insurance_data
```

```python
## Make hypothesis.. 
1. BMI will affect Insurance (Age inc - Charges inc)
    1. BMI - Smoker - cost - increase 
```

```python
sns.scatterplot(x = insurance_data['age'], y = insurance_data['charges'])
```

```python
sns.scatterplot(x = insurance_data['age'], y = insurance_data['charges'],hue = insurance_data['children'])
```

```python
sns.scatterplot(x = insurance_data['age'], y = insurance_data['charges'],hue = insurance_data['smoker'])
```

```python
#NOT A LEGIBLE WAY TO DO! JUST FOR EXAMPLE!
sns.scatterplot(x = insurance_data['age'], y = insurance_data['charges'],hue = insurance_data[['smoker','children']].apply(tuple,axis = 1))
```

```python
sns.scatterplot(x = insurance_data['bmi'], y = insurance_data['charges'])
```

```python
sns.scatterplot(x = insurance_data['bmi'], y = insurance_data['charges'],hue = insurance_data['smoker'])
```

```python
sns.swarmplot(x = insurance_data['smoker'], y = insurance_data['charges'])
```

```python
#Figure size setting
plt.figure(figsize = (20,10))

plt.title("Heatmap of all the columns for insurance charges")


insurance_corr = insurance_data.corr()
sns.heatmap(data=insurance_corr,annot = True,vmin = -1,vmax = 1,center=0)

plt.xlabel("columns Names")
```

```python
sns.pairplot(insurance_data)
```

```python
sns.lmplot(x = 'bmi', y = 'charges',hue = 'smoker',data =insurance_data)
```

```python
sns.regplot(x = 'bmi', y = 'charges',data =insurance_data)
```

```python
sns.displot(data = insurance_data['bmi'],kde = False)
```

```python
sns.displot(data = insurance_data['bmi'],kde = True)
```

```python
sns.kdeplot(data = insurance_data['bmi'],shade = True)
```

```python
sns.jointplot(x = insurance_data['bmi'],y = insurance_data['charges'],kind = 'kde')
```
