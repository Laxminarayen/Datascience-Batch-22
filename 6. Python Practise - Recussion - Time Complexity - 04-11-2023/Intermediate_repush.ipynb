```python
#Factorial of a number
#3! = 3*2*1
#4! = 4*3*2*1
#10! = 10*9*8*7*....1

#Used in Counting Theory Think of the example we saw in class - Selecting 5 board members from 200 people
#n!/(n-r)! r -> 5 (number of places)

#EDGE cases! -  0< n < 10000 !!!!!
0! = 1
```

```python
#Write this without recurssion
def factorial(num):
  fact = 1
  if (num==0)or(num ==1):
    return 1
  else:
    for i in range(num,1,-1): #Every element in the iterable is being visited in this loop..
      fact*=i
  return fact
```

```python
#Recurssion is a type of a func that calls itself
def fact_recurssion(n):
  if (n==1):
    return 1
  else:
    return n*fact_recurssion(n-1)
```

```python
count = 0
def count_substring(strng, sub_strng, count = count):
  if len(strng)==0:
    return 0
  if len(strng) < len(sub_strng):
    return 0
  else:
    if strng[:len(sub_strng)] == sub_strng:
      #print(strng[:len(sub_strng)])
      count+=1
    strng = strng[1:]
    return count_substring(strng,sub_strng,count)
```
